```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the path to the CIFAR-10 dataset
dataset_path = "C:\\Users\\aniks\\Onedrive\\Documents\\CIFAR-10-images-master"

# Function to convert an image to grayscale
def convert_to_grayscale(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    return gray_image

# Function to preprocess a folder of images
def preprocess_and_display_images(folder_path):
    # Get a list of all subfolders (e.g., airplane, automobile, bird)
    subfolders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]

    for subfolder in subfolders:
        subfolder_path = os.path.join(folder_path, subfolder)
        image_files = [f for f in os.listdir(subfolder_path) if f.endswith(".jpg")]

        # Display only the first image found in the subfolder
        if image_files:
            image_file = image_files[0]
            image_path = os.path.join(subfolder_path, image_file)
            image = cv2.imread(image_path)
            gray_image = convert_to_grayscale(image)

            # Save the grayscale image back to the same location
            cv2.imwrite(image_path, gray_image)

            # Show the grayscale image using matplotlib
            plt.imshow(gray_image, cmap='gray')
            plt.axis('off')
            plt.show()

# Preprocess the train and test folders along with their subfolders
preprocess_and_display_images(os.path.join(dataset_path, "train"))
preprocess_and_display_images(os.path.join(dataset_path, "test"))

print("Grayscale conversion completed.")
```
